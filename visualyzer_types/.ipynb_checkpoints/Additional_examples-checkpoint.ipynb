```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 3))

# Create circles
circle1 = patches.Circle((1, 1), radius=0.6, facecolor='black', edgecolor='black', linewidth=8)
ax.add_patch(circle1)

circle2 = patches.Circle((2.9, 1), radius=0.6, facecolor='black', edgecolor='black', linewidth=8)
ax.add_patch(circle2)

# Create connected lines between circles
y_l = 1
ax.plot([1.5, 2.3], [y_l, y_l], color='black', linewidth=5)

# Set the x and y limits
ax.set_xlim(0, 20)
ax.set_ylim(0, 4)

# Add => sign above the circles
ax.annotate("=>", xy=(1.95, 3), xytext=(1.95,3), fontsize=40, ha='center', va='center')

plt.axis("off")
# Show the plot
plt.show()
```

## Assign x and y to a numeric value:

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 6))  # Increase the figure height

# rhomb1
romb1 = patches.Polygon(
    [(1.2, 2), (1.8, 2.5), (2.4, 2), (1.8, 1.5)],
    facecolor='black', edgecolor='black', linewidth=2)
ax.add_patch(romb1)

ax.text(1, 2.7, 'x', fontsize=30, color='black', ha='center', va='center')

# rhomb2
romb2 = patches.Polygon(
    [(1.2, 0.5), (1.8, 1), (2.4, 0.5), (1.8, 0)],
    facecolor='black', edgecolor='black', linewidth=2)
ax.add_patch(romb2)

ax.text(1, 1, 'y', fontsize=30, color='black', ha='center', va='center')

# Adjust axis limits to ensure visibility
ax.set_xlim(0, 10)
ax.set_ylim(-1, 3)  

plt.axis("off")
# Show the plot
plt.show()




```

```python
# Example of a for loop with range, incrementing start value by 1
start_value = 3
end_value = 8

for i in range(start_value, end_value):
    start_value += end_value
    print(f"Current value: {start_value}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches  # Add this line
import numpy as np

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 6))  # Increase the figure height

# rhomb1
romb1 = patches.Polygon(
    [(1.2, 2), (1.8, 2.5), (2.4, 2), (1.8, 1.5)],
    facecolor='black', edgecolor='black', linewidth=2)
ax.add_patch(romb1)

ax.text(1, 2.7, 'x', fontsize=30, color='black', ha='center', va='center')

# rhomb2
romb2 = patches.Polygon(
    [(1.2, 0.5), (1.8, 1), (2.4, 0.5), (1.8, 0)],
    facecolor='black', edgecolor='black', linewidth=2)
ax.add_patch(romb2)

ax.text(1, 1, 'y', fontsize=30, color='black', ha='center', va='center')

# Adjust axis limits to ensure visibility
ax.set_xlim(0, 10)
ax.set_ylim(-1, 3)  

plt.axis("off")
# Show the plot
plt.show()

# for loop
# Create circle points
theta = np.linspace(0, 2*np.pi, 100)
r = 1  # radius of the circle

x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot the circle in black
plt.figure(figsize=(5, 5))
plt.plot(x, y, color='black')  # Change circle color to black

# Add arrows on the top and bottom with increased thickness and in black color
arrow_length = 0.6
arrow_prop_top = dict(arrowstyle='<-', shrinkA=0, shrinkB=0, linewidth=3, color='black')  # Black arrow
arrow_prop_bottom = dict(arrowstyle='<-', shrinkA=0, shrinkB=0, linewidth=3, color='black')  # Black arrow

# Arrow on the left
plt.annotate('', xy=(x[int(len(theta) * 0.48)], y[int(len(theta) * 0.48)]),
             xytext=(x[int(len(theta) * 0.53)], y[int(len(theta) * 0.53)]),
             arrowprops=arrow_prop_top)

# Arrow on the right
plt.annotate('', xy=(x[int(len(theta) * 0.95)], y[int(len(theta) * 0.95)]),
             xytext=(x[int(len(theta) * 0.99)], y[int(len(theta) * 0.99)]),
             arrowprops=arrow_prop_bottom)

# Create an axes instance
ax1 = plt.gca()

# Arrows with text
plt.annotate('', xy=(0, -1), xytext=(1, -1),
            arrowprops=dict(facecolor='black', arrowstyle='<-', linewidth=3))



# inside

romb3 = patches.Polygon(
    [(-0.25, 0.1), (-0.45, 0.3), (-0.65, 0.1), (-0.45, -0.1)],
    facecolor='black', edgecolor='black', linewidth=5)
ax1.add_patch(romb3)


ax1.text(-0.65, 0.4, 'x', fontsize=20, color='black', va='center', ha='center')

ax1.text(0.35, 0.4, 'y', fontsize=20, color='black', va='center', ha='center')

ax1.text(0.25, 0.15, '. . .', fontsize=30, color='black', va='center', ha='right')

# Create a polygon (romb2)
romb4 = patches.Polygon(
    [(0.35, 0.1), (0.55, 0.3), (0.75, 0.1), (0.55, -0.1)],
    facecolor='black', edgecolor='black', linewidth=5)
ax1.add_patch(romb4)



plt.axis("off")  # Turn off the axes
plt.grid(True)  # Turn off the grid
plt.show()
```
