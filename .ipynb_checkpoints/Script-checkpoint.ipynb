```python
#For loop
import matplotlib.pyplot as plt
import numpy as np

# Create circle points
theta = np.linspace(0, 2*np.pi, 100)
r = 1  # radius of the circle

x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot the circle in black
plt.figure(figsize=(3,3))
plt.plot(x, y, color='black')  # Change circle color to black

# Add arrows on the top and bottom with increased thickness and in black color
arrow_length = 0.6
arrow_prop_top = dict(arrowstyle='<-', shrinkA=0, shrinkB=0, linewidth=3, color='black')  # Black arrow
arrow_prop_bottom = dict(arrowstyle='<-', shrinkA=0, shrinkB=0, linewidth=3, color='black')  # Black arrow

# Arrow on the left
plt.annotate('', xy=(x[int(len(theta) * 0.48)], y[int(len(theta) * 0.48)]), 
             xytext=(x[int(len(theta) * 0.53)], y[int(len(theta) * 0.53)]), 
             arrowprops=arrow_prop_top)

# Arrow on the right
plt.annotate('', xy=(x[int(len(theta) * 0.95)], y[int(len(theta) * 0.95)]), 
             xytext=(x[int(len(theta) * 0.99)], y[int(len(theta) * 0.99)]), 
             arrowprops=arrow_prop_bottom)

# Create an axes instance
ax = plt.gca()


# Arrows with text
plt.annotate('', xy=(0, -1), xytext=(1, -1),
            arrowprops=dict(facecolor='black', arrowstyle='<-', linewidth=3))

plt.annotate('', xy=(0, -1), xytext=(1, -1.2),
            arrowprops=dict(facecolor='black', arrowstyle='<-', linewidth=3))

plt.axis('off')  # Turn off the axes
plt.grid(False)  # Turn off the grid
plt.title("For loop")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.lines import Line2D
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 3))

x = 1.5
y = 2

# Shapes of keys##############

#1-element
circle = Circle((x, y), 0.3, linewidth=1, edgecolor='black', facecolor='black')
ax.add_patch(circle)


#2-element
rhombus1 = patches.Polygon(
            [(1.5, 0.5), (1.2, 0.8), (1.5, 1.1), (1.8, 0.8)],
            facecolor='black', edgecolor='black', linewidth=5)
ax.add_patch(rhombus1)



####################################################

#Shapes of values###############
#1-element
circle2 = Circle((x+2, y), 0.3, linewidth=1, edgecolor='black', facecolor='black')
ax.add_patch(circle2)
circle3 = Circle((x+2.8, y), 0.3, linewidth=1, edgecolor='black', facecolor='black')
ax.add_patch(circle3)
ax.plot([3.5, 4], [y, y], color='black', linewidth=5)
######

#2-element
width=0.8
height=0.7

rectangle1 = patches.Rectangle((3.2, 0.5), width, height, linewidth=4, edgecolor='black', facecolor='none')
ax.add_patch(rectangle1)


######

#############################


# Line of keys and teeth###############
#1-element
line1 = Line2D([x + 0.2, x + 1.5], [y,y], color='black', linewidth=5)
ax.add_line(line1)
t1 = Line2D([x + 1.5, x + 1.5], [y - 0.25, y], color='black', linewidth=5)
t2 = Line2D([x + 0.4, x + 0.4], [y - 0.25, y], color='black', linewidth=5)
t3 = Line2D([x + 0.8, x + 0.8], [y - 0.25, y], color='black', linewidth=5)
ax.add_line(t1)
ax.add_line(t2)
ax.add_line(t3)
####

#2-element
y2=0.8
line2 = Line2D([x + 0.2, x + 1.5], [y2, y2], color='black', linewidth=5)
ax.add_line(line2)
t1_1 = Line2D([x + 1.5, x + 1.5], [y2 - 0.25, y2], color='black', linewidth=5)
t2_1 = Line2D([x + 0.4, x + 0.4], [y2 - 0.25, y2], color='black', linewidth=5)
t3_1 = Line2D([x + 0.8, x + 0.8], [y2 - 0.25, y2], color='black', linewidth=5)
ax.add_line(t1_1)
ax.add_line(t2_1)
ax.add_line(t3_1)
###



#############################################

# Set the x and y limits
ax.set_xlim(0, 15)
ax.set_ylim(0, 5)

plt.axis("off")
plt.show()
dictionary={'B':"ASDFGHJ", 4 :[1]}
```
